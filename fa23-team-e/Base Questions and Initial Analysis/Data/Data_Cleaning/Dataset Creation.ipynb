```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing the AQI dataset
AQI_file_path = '../../Data/AQI/2021_Avg_Aqi_Data_Cleaned.csv'
AQI = pd.read_csv(AQI_file_path)

#Importing the DP02 dataset
DP02_file_path = "../../Data/Census/DP02_Cleaned.csv"
DP02 = pd.read_csv(DP02_file_path)


#Importng the DP03 dataset
DP03_file_path = "../../Data/Census/DP03_Cleaned.csv"
DP03 = pd.read_csv(DP03_file_path)

#Importing the DP04 dataset
DP04_file_path = "../../Data/Census/DP04_Cleaned.csv"
DP04 = pd.read_csv(DP04_file_path)

#Importing the DP05 dataset
DP05_file_path = "../../Data/Census/DP05_Cleaned.csv"
DP05 = pd.read_csv(DP05_file_path)
```

```python
print(AQI.shape)
print(DP02.shape)
print(DP03.shape)
print(DP04.shape)
print(DP05.shape)
```

```python
# Standardizing the column names for zip code
AQI.rename(columns={AQI.columns[0]: 'Zipcode'}, inplace=True)
DP02.rename(columns={DP02.columns[0]: 'Zipcode'}, inplace=True)
DP03.rename(columns={DP03.columns[0]: 'Zipcode'}, inplace=True)
DP04.rename(columns={DP04.columns[0]: 'Zipcode'}, inplace=True)
DP05.rename(columns={DP05.columns[0]: 'Zipcode'}, inplace=True)

# Merging all datasets sequentially
merged_df = AQI.merge(DP02, on='Zipcode', how='inner')
merged_df = merged_df.merge(DP03, on='Zipcode', how='inner')
merged_df = merged_df.merge(DP04, on='Zipcode', how='inner')
merged_df = merged_df.merge(DP05, on='Zipcode', how='inner')

print(merged_df.shape)
merged_df.head()
```

```python
merged_df.to_csv('../../Data/Merged_AQI_DP/AQI_DP2_3_4_5.csv', index=False)
```
