```python
import pandas as pd
```

```python
final_merged_aqi = pd.read_csv('FinalMergedAQI.csv')
census_data = pd.read_csv('CensusData.csv')
```

```python
census_data.rename(columns={'Zip code': 'zip_code'}, inplace=True)
```

```python
# Merging the datasets on the 'zip_code' column, keeping only zip codes present in both
merged_final_census_corrected = final_merged_aqi.merge(census_data, on='zip_code', how='inner')
```

```python
merged_final_census_corrected.head()
```

```python
drop_cols = ['Unnamed: 0_x', 'Unnamed: 0_y']
merged_final_census_corrected.drop(columns=drop_cols)
```

```python
merged_final_census_corrected.to_csv('AQICensusMerged.csv')
```
