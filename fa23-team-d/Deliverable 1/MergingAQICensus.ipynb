```python
import pandas as pd
```

```python
aqi_avg_df = pd.read_csv('/Users/mithatkus/Desktop/BU/Fall23/CS506/Project/AQI data/CleanedYearlyAQI/2021AQIAvgbyZipcode.csv')
merged_census_df = pd.read_csv('/Users/mithatkus/Desktop/BU/Fall23/CS506/Project/2021Census/MergedCensus.csv')
```

```python
final_merged_df = pd.merge(merged_census_df, aqi_avg_df, left_on="Zip code", right_on="zip_code", how="left")
final_merged_df = final_merged_df.drop(columns=['zip_code'])
```

```python
zip_codes_to_remove = [2132, 2136, 2126]
final_merged_df = final_merged_df[~final_merged_df['Zip code'].isin(zip_codes_to_remove)]
```

```python
final_merged_df['Median price of owner occupied units (dollars)'] = final_merged_df['Median price of owner occupied units (dollars)'].str.replace(',', '')
final_merged_df['Median price of owner occupied units (dollars)'] = final_merged_df['Median price of owner occupied units (dollars)'].str.replace('+', '')
final_merged_df['Median price of owner occupied units (dollars)'] = final_merged_df['Median price of owner occupied units (dollars)'].replace('-', float('nan'))

# Convert the column to float
final_merged_df['Median price of owner occupied units (dollars)'] = final_merged_df['Median price of owner occupied units (dollars)'].astype(float)
```

```python
final_merged_df.isnull().sum()
```

```python
mean_value = final_merged_df['Median price of owner occupied units (dollars)'].mean()
final_merged_df['Median price of owner occupied units (dollars)'].fillna(mean_value, inplace=True)

final_merged_df = final_merged_df.dropna(subset=['PM2.5AQI'])

```

```python
final_merged_df.isnull().sum()
```

```python
column_to_drop = 'Unnamed: 0'
final_merged_df.drop(column_to_drop, axis=1)
```
