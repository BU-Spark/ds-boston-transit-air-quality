```python
import folium
import pandas as pd
from datetime import datetime

# Data
geojson_data = "/home/azad/Code/ds-boston-transit-air-quality/fa23-team-a/data/geojson/crb.geojson"

boston_sensors = pd.read_csv("~/Code/ds-boston-transit-air-quality/fa23-team-a/data/csv_files/purpleAirExtension.csv")

# Convert unix timestamp to date
boston_sensors['time_stamp'] = boston_sensors['time_stamp'].astype(int)
boston_sensors['time_stamp'] = pd.to_datetime(boston_sensors['time_stamp'], unit ='s')

# Mapping of sensor indices to places
sensor_places = {
    26219: 'Rosindale',
    28273: 'Brighton',
    29389: 'Cambridge'
}

# Add a new column 'place' based on the mapping
boston_sensors['place'] = boston_sensors['sensor_index'].map(sensor_places)

boston_sensors
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

boston_sensors = boston_sensors.sort_values(by='time_stamp')

sensor_26219 = boston_sensors[boston_sensors['sensor_index'] == 26219]
sensor_28273 = boston_sensors[boston_sensors['sensor_index'] == 28273]
sensor_29389 = boston_sensors[boston_sensors['sensor_index'] == 29389]

# Plotting for Sensor 26219, Rosindale
plt.figure(figsize=(12, 4))
plt.plot(sensor_26219['time_stamp'], sensor_26219['pm2.5_atm'], label='PM2.5 - Rosindale')
plt.plot(sensor_26219['time_stamp'], sensor_26219['pm10.0_atm'], label='PM10.0 - Rosindale')
plt.xlabel('Time')
plt.ylabel('PM Values')
plt.title('Rosindale - PM2.5 and PM10.0 Values Over Time')
plt.legend()
plt.show()

# Plotting for Sensor 28273, Brightton
plt.figure(figsize=(12, 4))
plt.plot(sensor_28273['time_stamp'], sensor_28273['pm2.5_atm'], label='PM2.5 - Brightton')
plt.plot(sensor_28273['time_stamp'], sensor_28273['pm10.0_atm'], label='PM10.0 - Brightton')
plt.xlabel('Time')
plt.ylabel('PM Values')
plt.title('Brightton - PM2.5 and PM10.0 Values Over Time')
plt.legend()
plt.show()

# Plotting for Sensor 29389, Cambridge
plt.figure(figsize=(12, 4))
plt.plot(sensor_29389['time_stamp'], sensor_29389['pm2.5_atm'], label='PM2.5 - Cambridge')
plt.plot(sensor_29389['time_stamp'], sensor_29389['pm10.0_atm'], label='PM10.0 - Cambridge')
plt.xlabel('Time')
plt.ylabel('PM Values')
plt.title('Cambridge - PM2.5 and PM10.0 Values Over Time')
plt.legend()
plt.show()
```
