### This script aggregates yearly airnow data for sites relevant to extension question

```python
import pandas as pd

airnow2020 = pd.read_csv('../data/airnow/airnow2020.csv')
airnow2021 = pd.read_csv('../data/airnow/airnow2021.csv')
airnow2022 = pd.read_csv('../data/airnow/airnow2022.csv')
airnow2023 = pd.read_csv('../data/airnow/airnow2023.csv')

airnow2020 = airnow2020[airnow2020['Status']!='Inactive']
airnow2021 = airnow2021[airnow2021['Status']!='Inactive']
airnow2022 = airnow2022[airnow2022['Status']!='Inactive']
airnow2023 = airnow2023[airnow2023['Status']!='Inactive']
```

```python
airnow2020_clean = airnow2020[['SiteName', 'Status', 'Latitude', 'Longitude',
       'PM10_AQI', 'PM25_AQI', 'NO2_AQI']]

airnow2021_clean = airnow2021[['SiteName', 'Status', 'Latitude', 'Longitude',
       'PM10_AQI', 'PM25_AQI', 'NO2_AQI']]

airnow2022_clean = airnow2022[['SiteName', 'Status', 'Latitude', 'Longitude',
       'PM10_AQI', 'PM25_AQI', 'NO2_AQI']]

airnow2023_clean = airnow2023[['SiteName', 'Status', 'Latitude', 'Longitude',
       'PM10_AQI', 'PM25_AQI', 'NO2_AQI']]
```

```python
airnow2020_agg = airnow2020_clean.groupby('SiteName').agg({
    'Latitude': 'first',  
    'Longitude': 'first',  
    'PM10_AQI': 'mean',
    'PM25_AQI': 'mean',
    'NO2_AQI': 'mean',
})

airnow2021_agg = airnow2021_clean.groupby('SiteName').agg({
    'Latitude': 'first',  
    'Longitude': 'first',   
    'PM10_AQI': 'mean',
    'PM25_AQI': 'mean',
    'NO2_AQI': 'mean',
})

airnow2022_agg = airnow2022_clean.groupby('SiteName').agg({
    'Latitude': 'first',  
    'Longitude': 'first',    
    'PM10_AQI': 'mean',
    'PM25_AQI': 'mean',
    'NO2_AQI': 'mean',
})

airnow2023_agg = airnow2023_clean.groupby('SiteName').agg({
    'Latitude': 'first',  
    'Longitude': 'first',  
    'PM10_AQI': 'mean',
    'PM25_AQI': 'mean',
    'NO2_AQI': 'mean',
})

airnow2020_agg['Year'] = 2020
airnow2021_agg['Year'] = 2021
airnow2022_agg['Year'] = 2022
airnow2023_agg['Year'] = 2023

airnow2020_agg
```

```python
airnow_yearly = pd.concat([airnow2020_agg, airnow2021_agg, airnow2022_agg, airnow2023_agg])
airnow_yearly = airnow_yearly.reset_index()
airnow_yearly = airnow_yearly[(airnow_yearly['SiteName']=="BOSTON-KENMORE") | (airnow_yearly['SiteName']=="Boston - Roxbury")].drop(columns=["PM10_AQI"])
airnow_yearly
```

```python
airnow_yearly.to_csv('../data/airnow_cleaned.csv', index=False)
```
