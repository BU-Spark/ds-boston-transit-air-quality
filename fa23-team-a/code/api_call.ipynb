## Extract CSV from AirNow API

This block of code gets the CSV containing AQI data from the year 2019 from AirNow

```python
import requests
import csv
import folium
import os
import sys
import urllib
from datetime import date
from datetime import datetime
from datetime import timedelta
from os.path import expanduser
from urllib.request import urlopen

# Function to convert data type into string compatible with API call
def date_to_str(d):
    s = ""
    s = d.strftime("%Y") + '-' + d.strftime("%m") + '-' + d.strftime("%d")
    return s

# Defining Parameters for API call
url = "https://www.airnowapi.org/aq/data/"
api_key = "INSERT API KEY"

start_date = datetime(2019, 1, 1)
end_date = datetime(2019,1,2)
    
options = {}
options["url"] = url
options["parameters"] = "PM25"
options["BBOX"] = "-71.271667,42.202941,-70.909495,42.475"
options["data_Type"] = "B"
options["format"] = "text/csv"
options["verbose"] = "0"
options["monitorType"] = "2"
options["includerawconcentrations"] = "0"
options["ext"] = "csv"
options["API_KEY"] = api_key

# directory where the output csv file will be
download_dir = expanduser("~/Code/ds-boston-transit-air-quality/fa23-team-a/data/csv_files")
# Name of CSV file
download_file_name = "AirNowAPI" + "." + options["ext"]

download_file = os.path.join(download_dir, download_file_name)

text = "Latitude,Longitude,UTC,Parameter,Concentration,Unit,AQI,Category\n"
with open(download_file, 'w') as f:
    f.write(text)


# Loop that makes an API call for every day of the year.
for x in range():
    
    options["start_Date"] = date_to_str(start_date)
    options["end_Date"] = date_to_str(end_date)
    # Construct URL
    REQUEST_URL = options["url"] \
                + "?startDate=" + options["start_Date"] + "T00"\
                + "&endDate=" + options["end_Date"] + "T00" \
                + "&parameters=" + options["parameters"] \
                + "&BBOX=" + options["BBOX"] \
                + "&dataType=" + options["data_Type"] \
                + "&format=" + options["format"] \
                + "&verbose=" + options["verbose"] \
                + "&monitorType=" + options["monitorType"] \
                + "&includerawconcentrations=" + options["includerawconcentrations"] \
                + "&API_KEY=" + options["API_KEY"]
    
    # Make get request
    api_data = urlopen(REQUEST_URL)
    
    # Write to CSV
    csv = api_data.read()
    with open(download_file, 'ab') as f:
        f.write(csv)

    # Increment days
    start_date += timedelta(days=1)
    end_date += timedelta(days=1)

    
print("Download File: ", download_file)
```
