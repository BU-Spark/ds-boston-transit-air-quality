```python
import pandas as pd
df = pd.read_csv('PLACES__ZCTA_Data__GIS_Friendly_Format___2023_release.csv')
df.columns
```

```python
df.head()
```

```python
# zipcodes = ['2134' '2116' '2116' '2135' '2129' '2122' '2128' '2115' '2127' '2136'
#  '2121' '2110' '2115' '2122' '2115' '2113' '2131' '2118' '2127' '2127'
#  '2116' '2114' '2136']

zipcodes = [2134, 2116, 2135, 2129, 2122, 2128, 2115, 2127, 2136, 2121, 2110, 2115, 2122, 2113, 2131, 2215, 2118, 2114, 2136, 2163, 2108, 2117, 2123, 2133, 2199, 2217, 2109, 2111, 2112, 2114, 2196, 2201, 2203, 2204, 2205, 2206, 2211, 2212, 2222, 2293, 2124, 2125, 2130, 2126, 2119, 2120, 2127, 2210, 2132, 2128, 2118, 2132, 2129, 2134, 2135, 2136, 2131, 2136, 2132]

filtered_df = df[df['ZCTA5'].isin(zipcodes)]
```

```python
print(df['ZCTA5'].head())
```

```python
filtered_df.head(12)
```

```python
columns_of_interest = ['ZCTA5',
    'CASTHMA_CrudePrev', 'COPD_CrudePrev',
    'CHD_CrudePrev', 'STROKE_CrudePrev', 'HIGHCHOL_CrudePrev',
    'GHLTH_CrudePrev', 'PHLTH_CrudePrev', 'MHLTH_CrudePrev',
    'CSMOKING_CrudePrev', 'BINGE_CrudePrev', 'LPA_CrudePrev',
    'OBESITY_CrudePrev', 'SLEEP_CrudePrev'
]

selected_columns_df = filtered_df[columns_of_interest]
selected_columns_df.rename(columns={'ZCTA5': 'ZipCode'}, inplace=True)

```

```python
selected_columns_df.head()
```

```python
summary = pd.read_csv('../Merged Data/result.csv')
merged_df = pd.merge(summary, selected_columns_df, on='ZipCode', how='inner')
merged_df.head()
```

```python
merged_df.to_csv('resulthealth.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = merged_df.copy()

health_columns = ['CASTHMA_CrudePrev', 'COPD_CrudePrev', 'CHD_CrudePrev', 'STROKE_CrudePrev', 'HIGHCHOL_CrudePrev']
pm25_column = 'AQI_PM2.5'

fig, ax1 = plt.subplots(figsize=(12, 6))

bottom_health = 0
legend_health = []

for column in health_columns:
    ax1.bar(data['Name'], data[column], bottom=bottom_health, label=column)
    bottom_health += data[column]
    legend_health.append(column)

ax1.set_xlabel('Neighborhood')
ax1.set_ylabel('Prevalence')
ax1.set_title('Health Conditions vs. PM2.5 by Neighborhood')
ax1.set_xticks(data['Name'])
ax1.set_xticklabels(data['Name'], rotation=90)

ax1.legend(legend_health, title="Health Conditions", loc="upper left")

ax2 = ax1.twinx()
ax2.plot(data['Name'], data[pm25_column], marker='o', color='red', label='PM2.5', linewidth=2)
ax2.set_ylabel('PM2.5', color='red')

ax2.legend(['PM2.5'], loc="upper right")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = merged_df.copy()

mbta_stops_column = 'MBTA_Stops_Count'
ozone_column = 'AQI_OZONE'

health_columns = ['CASTHMA_CrudePrev', 'COPD_CrudePrev', 'CHD_CrudePrev', 'STROKE_CrudePrev', 'HIGHCHOL_CrudePrev']

fig, ax1 = plt.subplots(figsize=(12, 6))

bottom_mbta_stops = 0
legend_mbta_stops = []

for column in health_columns:
    ax1.bar(data['Name'], data[column], bottom=bottom_mbta_stops, label=column)
    bottom_mbta_stops += data[column]
    legend_mbta_stops.append(column)

ax1.set_xlabel('Neighborhood')
ax1.set_ylabel('Prevalence')
ax1.set_title('Health Conditions vs. MBTA Stops by Neighborhood')
ax1.set_xticks(data['Name'])
ax1.set_xticklabels(data['Name'], rotation=90)

ax1.legend(legend_mbta_stops, title="Health Conditions", loc="upper left")

ax2 = ax1.twinx()
ax2.plot(data['Name'], data[mbta_stops_column], marker='o', color='blue', label='MBTA Stops', linewidth=2)
ax2.set_ylabel('Number of MBTA Stops', color='blue')

ax2.legend(['MBTA Stops'], loc="upper right")

ax3 = ax1.twinx()
ax3.plot(data['Name'], data[ozone_column], marker='s', color='green', label='Ozone', linewidth=2)
ax3.set_ylabel('Ozone Levels', color='green')
ax3.spines['right'].set_position(('outward', 60))

ax3.legend(['Ozone'], loc="upper right", bbox_to_anchor=(1,0.92))
plt.show()
```
