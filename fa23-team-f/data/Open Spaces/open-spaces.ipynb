```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df = pd.read_csv('open-spaces.csv')
```

```python
print(df)
```

```python
df.plot(x='District', y='Number of Open Spaces', kind='bar', width=0.8)
plt.xlabel('District')
plt.ylabel('Number of Open Spaces')
plt.title('Number of Open Spaces in each District')
# plt.xticks(rotation=45)
# plt.tight_layout()
plt.savefig('open-spaces-plot.png', bbox_inches='tight')
plt.show()
```
