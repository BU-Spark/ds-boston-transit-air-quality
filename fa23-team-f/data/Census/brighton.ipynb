```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame based on your data
data = {
    'Neighborhood and the Tracts within Brighton': [
        'Non-Hispanic White alone',
        'Non-Hispanic Black or African American alone',
        'Non-Hispanic Asian, Native Hawaiian and Pacific Islander alone',
        'Hispanic or Latino',
        'Others',
    ],
    'Population': [30965, 2379, 8046, 5993, 2220]
}
df = pd.DataFrame(data)

# Extract the column you want to visualize as a pie chart
labels = df['Neighborhood and the Tracts within Brighton']
sizes = df['Population']

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Race & Ethnicity in Brighton')
plt.savefig(f'pie_chart_brighton.png', bbox_inches='tight')
# Display the pie chart
plt.show()
```
