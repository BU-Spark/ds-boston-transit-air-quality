```python
import pandas as pd

df = pd.read_csv('mattapan-race.csv')

df = df.drop(0)

# Reset the index if needed
df = df.reset_index(drop=True)

df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame based on your data
data = {
    'Neighborhood and the Tracts within Mattapan': [
        'Non-Hispanic White alone',
        'Non-Hispanic Black or African American alone',
        'Non-Hispanic Asian, Native Hawaiian and Pacific Islander alone',
        'Hispanic or Latino',
        'Others',
    ],
    'Population': [1389, 19418, 450, 4070, 1611]
}
df = pd.DataFrame(data)

# Extract the column you want to visualize as a pie chart
labels = df['Neighborhood and the Tracts within Mattapan']
sizes = df['Population']

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Race & Ethnicity in Mattapan')
plt.savefig(f'pie_chart_mattapan.png', bbox_inches='tight')
# Display the pie chart
plt.show()
```
