```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df = pd.read_csv('suffolk-race.csv')
df
```

```python
df['Percent'] = df['Percent'].apply(lambda x: float(x.rstrip('%')))

plt.figure(figsize=(8, 8))  
plt.pie(df['Percent'], labels=df['Demographic'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Race in Suffolk County')
plt.savefig('race-suffolk-plot.png', bbox_inches='tight')
plt.show()
```
