```python
import pandas as pd

df = pd.read_csv('../SocialVulnerability//Climate_Ready_Boston_Social_Vulnerability.csv')

df = df.groupby('Name').sum()

df_sum = pd.read_csv('result.csv')
df_air = df_sum[['Name', 'AQI']]

merged_data = pd.merge(df, df_air, on='Name', how='inner')
merged_data.head()
newdf = merged_data[['Name', 'TotDis', 'TotChild', 'OlderAdult', 'Low_to_No' ,'LEP', 'POC2', 'MedIllnes', 'AQI']]
newdf.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


data = newdf.copy()

data['Population'] = data['TotDis'] + data['TotChild'] + data['OlderAdult'] + data['Low_to_No'] + data['LEP'] + data['POC2'] + data['MedIllnes']
stacked_columns = ['TotDis', 'TotChild', 'OlderAdult', 'Low_to_No', 'LEP', 'POC2', 'MedIllnes']

selected_columns = data[['Name', 'Population', 'AQI'] + stacked_columns]
stacked_columns = ['TotDis', 'TotChild', 'OlderAdult', 'Low_to_No', 'LEP', 'POC2', 'MedIllnes']

fig, ax1 = plt.subplots(figsize=(12, 6))

bottom = 0
legend_labels = []

for column in stacked_columns:
    ax1.bar(data['Name'], data[column], bottom=bottom, label=column)
    bottom += data[column]
    legend_labels.append(column)

ax1.set_xlabel('Neighborhood')
ax1.set_ylabel('Population')
ax1.set_title('Social Vulnerability Factors by Neighborhood Population vs. AQI')
ax1.set_xticks(data['Name'])
ax1.set_xticklabels(data['Name'], rotation=90)

ax1.legend(legend_labels, title="Categories", loc="center left", bbox_to_anchor=(1.1, 0.5), fontsize='small')

ax2 = ax1.twinx()
aqi_data = data['AQI']
ax2.plot(data['Name'], aqi_data, marker='o', color='red', label='AQI', linewidth=2)
ax2.set_ylabel('AQI', color='red')

ax2.legend(['AQI'], loc="upper left", bbox_to_anchor=(0, 0.5))

plt.show()

selected_columns.head(10)
```
