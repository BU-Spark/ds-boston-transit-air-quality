```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

mdata = pd.read_csv("result.csv")

mdata['Total_PM'] = mdata['PM2.5'] + mdata['PM10']

plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='Total_PM', data=mdata, color='green')
plt.xlabel('Neighborhood')
plt.ylabel('Total PM (PM2.5 + PM10)')
plt.title('Total PM (PM2.5 + PM10) per Neighborhood')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

mdata = pd.read_csv("result.csv")

plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='AQI', data=mdata, color='orange')
plt.xlabel('Neighborhood')
plt.ylabel('AQI')
plt.title('Total AQI per Neighborhood')
plt.xticks(rotation=90)
plt.show()
```

```python
df = pd.read_csv('result.csv')

correlation = df['Blue_Bikes_Docks_Count'].corr(df['AQI'])

# Print the correlation coefficient
print(f"Correlation coefficient: {correlation}")

plt.scatter(df['Blue_Bikes_Docks_Count'], df['AQI'])
plt.title('Scatter Plot of Blue_Bikes_Docks_Count vs AQI')
plt.xlabel('Blue_Bikes_Docks_Count')
plt.ylabel('AQI')
plt.grid(True)
plt.show()

sns.regplot(x='Blue_Bikes_Docks_Count', y='AQI', data=df)
plt.title('Scatter Plot with Regression Line of Blue_Bikes_Docks_Count vs AQI')
plt.show()
```
