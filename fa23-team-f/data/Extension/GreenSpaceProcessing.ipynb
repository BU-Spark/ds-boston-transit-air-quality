```python
dfy = pd.read_csv('AQIthroughNeighborhood.csv')
dfy
```

```python
dfx = pd.read_csv('merged_data.csv')
dfx
```

```python
X = dfx['Name'].drop_duplicates().tolist()
Y = dfy['Name'].drop_duplicates().tolist()

newList = []
for x in X:
    if x in Y:
        newList.append(x)

for y in Y:
    if y in X and y not in newList:
        newList.append(y)

newList
```

```python
Xvalues = []
Yvalues = []
for n in newList:
    Xvalues.append(dfx.loc[dfx['Name'] == n]['PARK_COUNT'])
    Yvalues.append(dfy.loc[dfy['Name'] == n]['AQI'])
```

```python
import matplotlib.pyplot as plt

plt.scatter(Xvalues, Yvalues)
```
