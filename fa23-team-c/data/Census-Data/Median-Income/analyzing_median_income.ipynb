```python
import pandas as np 
import pandas as pd

# Load the data from CSV files
df_2019 = pd.read_csv('Median-Income2019.csv')
df_2020 = pd.read_csv('Median-Income2020.csv')
df_2021 = pd.read_csv('Median-Income2021.csv')

# Display the first few rows to confirm successful loading
df_2019.head()
```

```python
# drop rows with nan values
df_2019.dropna(inplace=True)
df_2020.dropna(inplace=True)
df_2021.dropna(inplace=True)
```

```python
zip_codes = set(col.split('!!')[0][-5:] for col in df_2019.columns if 'ZCTA5' in col)
metrics = []
cols = set(f'ZCTA5 {zip_code}!!{metric}' for zip_code in zip_codes for metric in metrics)
```

```python
def avg_all_rows(df):
    """Return the average of all rows in the dataframe."""
    return df.mean(axis=1)

df_2019 = avg_all_rows(df_2019)
```

```python
def get_mean_income(df):
    zip_codes = set(col.split('!!')[0][-5:] for col in df_2019.columns if 'ZCTA5' in col)
    metrics = [ 'Mean income (dollars)!!Estimate']
    rows = []
    for _, row in df_2019.iterrows():
        row_data = {}
        for zip_code in zip_codes:
            for metric in metrics:
                pattern = f'ZCTA5 {zip_code}!!{metric}'
                for col in df_2019.columns:
                    if pattern in col:
                        # Assign the corresponding value from the current row
                        row_data[f'{zip_code} {metric}'] = row[col]
                        break
        rows.append(row_data)

    # Create a new DataFrame from the rows list
    new_df = pd.DataFrame(rows)

    # Check the result
    return new_df

df_2019 = get_mean_income(df_2019)
df_2020 = get_mean_income(df_2020)
df_2021 = get_mean_income(df_2021)
```

```python
def process_df(df):
    # List of unique zip codes
    zip_codes = ['02118', '02119', '02121', '02125', '02215']

    # List of desired substrings
    substrings = [
        'Number!!Estimate', 
        'Number!!Margin of Error', 
        'Percent Distribution!!Estimate', 
        'Percent Distribution!!Margin of Error', 
        'Mean income (dollars)!!Estimate', 
        'Mean income (dollars)!!Margin of Error'
    ]

    # Initialize an empty DataFrame
    processed_df = pd.DataFrame()

    # Iterate over each zip code and substring
    for zip_code in zip_codes:
        for substring in substrings:
            # Construct the full string to match
            full_string = f'ZCTA5 {zip_code}!!{substring}'

            # Find the column in the original DataFrame that matches this string
            matching_columns = [col for col in df.columns if full_string in col]

            # Check if there is exactly one matching column
            if len(matching_columns) == 1:
                # Add this column to the new DataFrame with a renamed column
                processed_df[f'{zip_code} {substring}'] = df[matching_columns[0]]
    
    return processed_df

df_2019 = process_df(df_2019)
df_2019.head()
```
