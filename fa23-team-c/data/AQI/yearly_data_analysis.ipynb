```python
def drop_unneeded_cols(df):
    df = df.drop(["StateName", "CountryCode", "DataSource", "ReportingArea_PipeDelimited", "ValidTime", "GMTOffset", "Elevation", "Longitude", "Latitude", "EPARegion", "AQSID", "Status"], axis=1)
    return df

df_2020 = drop_unneeded_cols(df_2020)
df_2021 = drop_unneeded_cols(df_2021)
df_2022 = drop_unneeded_cols(df_2022)
df_2023 = drop_unneeded_cols(df_2023)
```

```python
numberic_col = ['OZONE_AQI', 'PM10_AQI', 'PM25_AQI', 'NO2_AQI']
def fill_nan(df):
    # fill na with mean value of the column
    df = df.fillna(df[numberic_col].mean())
    return df

df_2020 = fill_nan(df_2020)
df_2021 = fill_nan(df_2021)
df_2022 = fill_nan(df_2022)
df_2023 = fill_nan(df_2023)
```

```python
# 2020 and 2021 has all na for PM10 -> replace with 0
df_2020["PM10_AQI"] = df_2020["PM10_AQI"].fillna(0)
df_2021["PM10_AQI"] = df_2021["PM10_AQI"].fillna(0)
```

```python
def create_avgAQI(df, interested_col, numberic_col):
    df = df[interested_col]
    df = df.groupby(['ValidDate', 'SiteName'])[numberic_col].mean()

    # create a new col call avg_AQI that takes the average of three AQI columns
    df['avg_AQI'] = df[numberic_col].mean(axis=1)
    return df

numberic_col = ['OZONE_AQI', 'PM25_AQI', 'NO2_AQI']
interested_col = ['SiteName', 'ValidDate'] + numberic_col
df_2020 = create_avgAQI(df_2020, interested_col, numberic_col)
df_2021 = create_avgAQI(df_2021, interested_col, numberic_col)

numberic_col_latest = ['OZONE_AQI', 'PM25_AQI', 'NO2_AQI', 'PM10_AQI']
interested_col_latest = ['SiteName', 'ValidDate'] + numberic_col_latest
df_2022 = create_avgAQI(df_2022, interested_col_latest, numberic_col_latest)
df_2023 = create_avgAQI(df_2023, interested_col_latest, numberic_col_latest)

df_2022.head(15)
```

```python
def ret_validDate_col(df):
    df = df.reset_index()
    df['ValidDate'] = pd.to_datetime(df['ValidDate'])
    df['ValidDate'] = df['ValidDate'].apply(lambda x: x.timetuple().tm_yday)
    return df

df_2020 = ret_validDate_col(df_2020)
df_2021 = ret_validDate_col(df_2021)
df_2022 = ret_validDate_col(df_2022)
df_2023 = ret_validDate_col(df_2023)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.dates import date2num

def plot_daily_per_year(df):
    plt.figure(figsize=(8, 6))
    plt.xticks(rotation=45)

    # Create the lineplot
    lineplot = sns.lineplot(x='ValidDate', y='avg_AQI', hue='SiteName', data=df)

    # Set x-axis limits
    plt.xlim(1, 366)

    # Adjust line opacity
    for line in lineplot.lines:
        line.set_alpha(0.5)

    plt.show()

plot_daily_per_year(df_2020)
plot_daily_per_year(df_2021)
plot_daily_per_year(df_2022)
plot_daily_per_year(df_2023)
```

```python
# Function to assign quarter based on day of the year
def assign_quarter(day):
    if day <= 90:
        return 'Quarter 1'
    elif day <= 181:
        return 'Quarter 2'
    elif day <= 273:
        return 'Quarter 3'
    else:
        return 'Quarter 4'

# Function to process each DataFrame
def process_df(df, year, exclude_q4=False):
    df['Quarter'] = df['ValidDate'].apply(assign_quarter)
    if exclude_q4 and year == 2023:
        df = df[df['Quarter'] != 'Quarter 4']
    quarterly_avg = df.groupby('Quarter')['avg_AQI'].mean().reset_index()
    quarterly_avg['Year'] = year
    return quarterly_avg

# Process each DataFrame
df_2020_q = process_df(df_2020, 2020)
df_2021_q = process_df(df_2021, 2021)
df_2022_q = process_df(df_2022, 2022)
df_2023_q = process_df(df_2023, 2023, exclude_q4=True)  # Exclude Quarter 4 for 2023

# Combine the DataFrames
combined_df = pd.concat([df_2020_q, df_2021_q, df_2022_q, df_2023_q])

plt.figure(figsize=(10, 6))

# Plotting
sns.lineplot(data=combined_df, x='Quarter', y='avg_AQI', hue='Year', marker='o')

plt.xlabel('Quarter')
plt.ylabel('Average AQI')
plt.title('Quarterly Average AQI (2020-2023)')
plt.tight_layout()
plt.savefig("../results/Quarterly Average AQI (2020-2023)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Function to assign quarter based on day of the year
def assign_quarter(day):
    if day <= 90:
        return 'Quarter 1'
    elif day <= 181:
        return 'Quarter 2'
    elif day <= 273:
        return 'Quarter 3'
    else:
        return 'Quarter 4'

# Function to process each DataFrame
def process_df(df, year):
    df['Quarter'] = df['ValidDate'].apply(assign_quarter)
    df['Year-Quarter'] = df['Quarter'] + '-' + str(year)
    quarterly_avg = df.groupby(['Year-Quarter', 'SiteName'])['avg_AQI'].mean().reset_index()
    return quarterly_avg

# Process each DataFrame
df_2020_q = process_df(df_2020, 2020)
df_2021_q = process_df(df_2021, 2021)
df_2022_q = process_df(df_2022, 2022)
df_2023_q = process_df(df_2023, 2023)

# Exclude Quarter 4 for 2023
df_2023_q = df_2023_q[~df_2023_q['Year-Quarter'].str.contains('Quarter 4-2023')]

# Combine the DataFrames
combined_df = pd.concat([df_2020_q, df_2021_q, df_2022_q, df_2023_q])

plt.figure(figsize=(12, 6))

# Plotting
sns.lineplot(data=combined_df, x='Year-Quarter', y='avg_AQI', hue='SiteName', marker='o')

plt.xticks(rotation=45)
plt.xlabel('Quarter-Year')
plt.ylabel('Average AQI')
plt.title('Quarterly Average AQI per Site (2020-2023)')
plt.tight_layout()
plt.savefig("../results/Quarterly Average AQI per Site (2020-2023)")
plt.show()
```
