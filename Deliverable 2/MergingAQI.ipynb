```python
import pandas as pd
```

```python
data_2017 = pd.read_csv('2017_daily_AQI_data.csv')
data_2018 = pd.read_csv('2018_daily_AQI_data.csv')
data_2019 = pd.read_csv('2019_daily_AQI_data.csv')
data_2021 = pd.read_csv('2021_daily_AQI_data.csv')
data_2022 = pd.read_csv('2022_daily_AQI_data.csv')
```

```python
# Preprocess each dataset to remove rows with NaN values
data_2017.dropna(inplace=True)
data_2018.dropna(inplace=True)
data_2019.dropna(inplace=True)
data_2021.dropna(inplace=True)
data_2022.dropna(inplace=True)
```

```python
merged_data = data_2017.merge(data_2018, on='zip_code', how='inner')
merged_data = merged_data.merge(data_2019, on='zip_code', how='inner')
merged_data = merged_data.merge(data_2021, on='zip_code', how='inner')
merged_data = merged_data.merge(data_2022, on='zip_code', how='inner')
```

```python
years = [2017, 2018, 2019, 2021, 2022]
final_data = pd.DataFrame()
```

```python
# Adjusting the code to use pd.concat instead of append
final_data_concat = pd.concat(
    [merged_data.filter(regex=f'(^zip_code|_{year}$)').assign(year=year).rename(columns=lambda x: x.replace(f'_{year}', ''))
     for year in years]
).sort_values(by=['zip_code', 'year']).reset_index(drop=True)

# Display the first 15 rows of the DataFrame to show the first three zip codes
final_data_concat.head(15)
```

```python
final_data_concat.to_csv('FinalMergedAQI.csv')
```
